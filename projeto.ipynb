## importacao de dados

```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

## 1. Análise do faturamento

```python
lojas = {"Loja 1": loja, "Loja 2": loja2, "Loja 3": loja3, "Loja 4": loja4}

faturamento = {nome: df['Preço'].sum() for nome, df in lojas.items()}
plt.bar(faturamento.keys(), faturamento.values())
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.show()
```

## 2. Vendas por Categoria

```python
for nome, df in lojas.items():
    categorias = df['Categoria do Produto'].value_counts().head(5)
    categorias.plot(kind='bar', title=f'Top 5 Categorias - {nome}')
    plt.xlabel('Categoria')
    plt.ylabel('Quantidade de Vendas')
    plt.show()
```

## 3. Média de Avaliação das Lojas

```python
avaliacao_media = {nome: df['Avaliação da compra'].mean() for nome, df in lojas.items()}
plt.barh(list(avaliacao_media.keys()), list(avaliacao_media.values()), color='orange')
plt.title('Avaliação Média dos Clientes')
plt.xlabel('Nota Média')
plt.show()
```

## 4. Produtos Mais e Menos Vendidos

```python
for nome, df in lojas.items():
    top_produtos = df['Produto'].value_counts().head(5)
    plt.pie(top_produtos.values,
            labels=top_produtos.index,
            autopct='%1.1f%%',
            startangle=90)
    plt.title(f'Top 5 Produtos Mais Vendidos - {nome}')
    plt.show()
```

```python
for nome, df in lojas.items():
    menos_vendidos = df['Produto'].value_counts().sort_values(ascending=True).tail(5)
    plt.pie(menos_vendidos.values,
            labels=menos_vendidos.index,
            autopct='%1.1f%%',
            startangle=90)
    plt.title(f'Top 5 Produtos Menos Vendidos - {nome}')
    plt.show()
```

## 5. Frete Médio por Loja


```python
frete_medio = {nome: df['Frete'].mean() for nome, df in lojas.items()}

plt.plot(list(frete_medio.keys()), list(frete_medio.values()), marker='o', color='green')
plt.title('Comparativo do Frete Médio por Loja')
plt.xlabel('Loja')
plt.ylabel('Valor Médio do Frete (R$)')
plt.grid(True)
plt.show()
```
